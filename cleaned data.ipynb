```python
#Loading the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("summarymeasures.csv")
df.head()
```

```python
#Dropping the unwanted columns
columns_to_remove=['CI_Min_Unhealthy_Days','CI_Max_Unhealthy_Days']
df=df.drop(columns=columns_to_remove)
df
```

```python
df.dtypes
```

```python
#Transforming the datatypes
df['Health_Status']=df['Health_Status'].round(0).astype(int)
df['CI_Min_Health_Status']=df['CI_Min_Health_Status'].round(0).astype(int)
df['CI_Max_Health_Status']=df['CI_Max_Health_Status'].round(0).astype(int)
df['Unhealthy_Days']=df['Unhealthy_Days'].round(0).astype(int)
df
```

```python
#Replacing the negative values with NaN
df[df['Health_Status']<0]=np.nan
df[df['CI_Min_Health_Status']<0]=np.nan
df[df['CI_Max_Health_Status']<0]=np.nan
df[df['Unhealthy_Days']<0]=np.nan
df
```

```python
columns_to_convert=['index','State_FIPS_Code','County_FIPS_Code']
df[columns_to_convert]=df[columns_to_convert].apply(pd.to_numeric, errors='coerce', downcast='integer')
df
```

```python
#Dealing with Null Values
df=df.fillna('')
df
```

```python
#Dropping Rows
df=df.dropna(how='any',axis=0)
df=df.reset_index(drop=True)
df
```

```python
#Renaming the column
df.rename(columns={'CHSI_State_Abbr':'State_Abbr'},inplace=True)
df
```

```python
df=df.drop_duplicates()
df
```

```python
#Creating a calculated field
df['Min_Unhealthy_Days'] = pd.to_numeric(df['Min_Unhealthy_Days'], errors='coerce')
df['Max_Unhealthy_Days'] = pd.to_numeric(df['Max_Unhealthy_Days'], errors='coerce')
df['Avg_Unhealthy_Days'] = (df['Min_Unhealthy_Days'] + df['Max_Unhealthy_Days']) / 2
df

```

```python
#Handling the outliers
import pandas as pd
df['US_Health_Status'] = pd.to_numeric(df['US_Health_Status'], errors='coerce')
threshold = 100
df['US_Health_Status'] = df['US_Health_Status'].where(df['US_Health_Status'] <= threshold, df['US_Health_Status'].median())
df
```
